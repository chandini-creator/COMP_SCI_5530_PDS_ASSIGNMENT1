```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df_cleaned = pd.read_csv("processed_studentsPerformance_data.csv")
```

1. Histogram - Math Score Distribution

```python
plt.figure(figsize=(8,6))
sns.histplot(df_cleaned["math score"], bins=10, kde=True, color="blue")
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.savefig("visualizations_1.png")
plt.show()
```

2. Bar Chart - Average Math Score by Gender

```python
plt.figure(figsize=(8,6))
sns.barplot(x="gender", y="math score", data=df_cleaned, estimator=sum,hue="gender", palette="coolwarm")
plt.title("Average Math Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Math Score")
plt.savefig("visualizations_2.png")
plt.show()
```

3. Scatter Plot - Math Score vs. Reading Score

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="math score", y="reading score", hue="gender", data=df_cleaned, palette="deep")
plt.title("Math Score vs. Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.savefig("visualizations_3.png")
plt.show()
```


4. Line Plot - Math Score Trend Across Different Parental Education Levels

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="parental level of education", y="math score", data=df_cleaned, marker='o',  color="purple")
plt.xticks(rotation=45)
plt.title("Math Score Trend Across Different Parental Education Levels")
plt.xlabel("Parental Education Level")
plt.ylabel("Math Score")
plt.savefig("visualizations_4.png")
plt.show()
```


5. Pie Chart - Distribution of Lunch Types

```python

plt.figure(figsize=(8,6))
df_cleaned["lunch"].value_counts().plot.pie(autopct='%1.1f%%', colors=["skyblue", "lightcoral"], startangle=90)
plt.title("Distribution of Lunch Types")
plt.ylabel("")
plt.savefig("visualizations_5.png")
plt.show()
```
